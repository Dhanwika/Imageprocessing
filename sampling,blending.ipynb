```python
#upsampling
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('lion.jpg')
cv2.imshow("image before upsampling",img)
img=cv2.pyrUp(img)
cv2.imshow("up sampling",img)
plt.imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#DOWN SAMPLING
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('lion.jpg')
cv2.imshow("image before downsampling",img)
img=cv2.pyrDown(img)
cv2.imshow("DOWN sampling",img)
plt.imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#adding,subtracting,multiplying,dividing two images
import cv2
img1=cv2.imread('star-1-300x168.jpg')
img2=cv2.imread('dot-300x168.jpg')
add=img1+img2
sub=img2-img1
mul=img1*img2
div=img2/img1
cv2.imshow("add",add)
cv2.imshow("sub",sub)
cv2.imshow("mul",mul)
cv2.imshow("div",div)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#BLENDING TWO IMAGES
import cv2
f1 = cv2.imread('star-1-300x168.jpg', 1)
f2 = cv2.imread('dot-300x168.jpg', 1)
img = cv2.addWeighted(f1, 0.3, f2, 0.7, 0)
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
