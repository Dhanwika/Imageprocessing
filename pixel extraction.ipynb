```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread,imshow
```

```python
#loading the image
image1=imread("img1.jfif")
imshow(image1)
```

```python
#converting the original image to grayscale
image2=imread("img1.jfif",as_gray=True)
imshow(image2)
```

```python
#shape of the image
print(image1.shape)
print(image2.shape)
```

```python
#size of the image
print(image1.size)
print(image2.size)
```

```python
#feature extraction

#pixel features
pixel_features1=np.reshape(image2,(159*318))
pixel_features1.shape,pixel_features1
```

```python
pixel_features2=np.reshape(image1,(159*318*3))
pixel_features2.shape,pixel_features2
```

```python
#variance of all pixels
from PIL import Image,ImageStat
im=Image.open("img1.jfif")
stat=ImageStat.Stat(im)
print(stat.var)
```
