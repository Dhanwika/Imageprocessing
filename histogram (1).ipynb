```python
#IMAGE CONTRAST
from PIL import Image, ImageEnhance
im = Image.open("img2.jfif")
#image brightness enhancer
enhancer = ImageEnhance.Contrast(im)
#conrast factor=1 for original image
#contrast factor is less than one for decreasing the contrast ofthe image and vice-versa
factor = 3 #increase contrast 
im_output = enhancer.enhance(factor)
im_output
```

```python
#BITWISE OPERATIONS
import numpy as np
import cv2
# draw a rectangle
rectangle = np.zeros((300, 300), dtype="uint8")
cv2.rectangle(rectangle, (25, 25), (275, 275), 255, -1)
cv2.imshow("Rectangle", rectangle)
# draw a circle
circle = np.zeros((300, 300), dtype = "uint8")
cv2.circle(circle, (150, 150), 150, 255, -1)
cv2.imshow("Circle", circle)
bitwiseAnd = cv2.bitwise_and(rectangle, circle)
cv2.imshow("AND", bitwiseAnd)
cv2.waitKey(0)
bitwiseOr = cv2.bitwise_or(rectangle, circle)
cv2.imshow("OR", bitwiseOr)
cv2.waitKey(0)
bitwiseXor = cv2.bitwise_xor(rectangle, circle)
cv2.imshow("XOR", bitwiseXor)
cv2.waitKey(0)
bitwiseNot = cv2.bitwise_not(circle)
cv2.imshow("NOT", bitwiseNot)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#BITWISE OPERATIONS AND,OR,NOT,XOR
import numpy as np
import cv2
f1 = cv2.imread('star-1-300x168.jpg', 1)
f2 = cv2.imread('dot-300x168.jpg', 1)
#BITWISE AND
bitwiseAnd = cv2.bitwise_and(f1, f2)
cv2.imshow("AND", bitwiseAnd)
cv2.waitKey(0)
cv2.destroyAllWindows()
#BITWISE OR
bitwiseOr = cv2.bitwise_or(f1, f2)
cv2.imshow("OR", bitwiseOr)
cv2.waitKey(0)
bitwiseXor = cv2.bitwise_xor(f1, f2)
cv2.imshow("XOR", bitwiseXor)
cv2.waitKey(0)
bitwiseNot = cv2.bitwise_not(f2)
cv2.imshow("NOT", bitwiseNot)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
f1 = cv2.imread('1-500x250-3.jpg', 1)
img = cv2.addWeighted(f1, 3, f1, 2, 10)
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img=cv2.imread('img2.jfif',0)
plt.hist(img.ravel(),256,[0,256])
plt.show()
plt.savefig('hist.png')
equ=cv2.equalizeHist(img)
res=np.hstack((img,equ))
cv2.imshow("Equalized image",res)
cv2.imwrite("Equalized image.png",res)
plt.hist(res.ravel(),256,[0,256])
plt.show()
plt.savefig('equal_hist.png')
cv2.waitKey()
cv2.destroyAllWindows()
```
